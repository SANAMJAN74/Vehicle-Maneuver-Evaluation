```python
# Import necessary libraries
from random import shuffle
import numpy as np
import pandas as pd
from scipy.special import logsumexp
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
import more_itertools
import random
import matplotlib.pyplot as plt
from random import shuffle
import pickle
from sklearn.metrics import confusion_matrix, precision_recall_fscore_support
import seaborn as sns
from scipy.special import logsumexp
from scipy.stats import multivariate_normal

# Define the file paths
LEFT_MANEUVER_FILE = 'Maneuvers/Left_Maneuvers.csv'
RIGHT_MANEUVER_FILE = 'Maneuvers/Right_Maneuvers.csv'
STRAIGHT_MANEUVER_FILE = 'Maneuvers/Straight_Maneuvers.csv'

def data_preparation(file_path):
    """
    Preprocess the maneuver data from a CSV file.
    
    This function reads the CSV file, parses the data into separate series based on headers,
    converts to numpy arrays, trims each series to rows 16:61 (45 data points), and creates
    a list of pandas DataFrames for each series.
    
    Parameters:
    file_path (str): Path to the CSV file.
    
    Returns:
    list: List of pandas DataFrames, each representing a series.
    """
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    # Split lines into lists, handling commas
    raw_data = [line.strip().split(',') for line in lines]
    
    # Initialize data structure: list of lists for each series
    data_series = [[]]
    count = 0
    for row in raw_data:
        if row[0] == 'Lateral_Position':
            data_series.append([])
            count += 1
            continue
        data_series[count].append(row)
    
    # Number of series
    num_series = len(data_series)
    print(f"Number of Series in {file_path} = {num_series}")
    
    # Convert each series to a numpy array of floats
    for j in range(num_series):
        data_series[j] = np.array([[float(k) for k in data] for data in data_series[j]])
    
    # Fixed length for trimming (based on user's code: rows 16:61, which is 45 points)
    trim_length = 45  # From index 16 to 60 inclusive: 60-16+1=45
    
    # Create list of DataFrames
    list_of_dfs = []
    for j in range(num_series):
        # Trim the data
        trimmed_data = data_series[j][16:16 + trim_length]
        df = pd.DataFrame(trimmed_data, columns=['Lateral_Position', 'Velocity', 'Acceleration'])
        list_of_dfs.append(df)
    
    print(f"Shape of a sample series: {list_of_dfs[0].shape if list_of_dfs else 'No data'}")
    
    return list_of_dfs

def plot_series_state(data_list, state, title='', xlab='Time (ms)', ylab='', axis=None, color='maroon'):
    """
    Plot the specified state from multiple series overlaid on a single plot.
    
    Parameters:
    data_list (list): List of pandas DataFrames.
    state (int): Column index (0: Lateral_Position, 1: Velocity, 2: Acceleration).
    title (str): Plot title.
    xlab (str): X-axis label.
    ylab (str): Y-axis label.
    axis (list): Axis limits [xmin, xmax, ymin, ymax].
    color (str): Line color for plots.
    """
    if not data_list:
        print("No data to plot.")
        return
    if axis is None:
        axis = [-1, 45, 0, 15]  # Default based on user's example
    
    fig = plt.figure()
    plt.axis(axis)
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.title(title)
    
    num_series = len(data_list)
    all_states = []  # To compute average if needed
    
    for i in range(num_series):
        s = data_list[i].iloc[:, state].values  # Extract the state column as array
        plt.plot(range(len(s)), s, color=color)
        all_states.append(s)
    
    plt.show()

# Load and preprocess all datasets
print("Preprocessing Straight Maneuvers...")
l_k = data_preparation(STRAIGHT_MANEUVER_FILE)

print("Preprocessing Left Maneuvers...")
l_c_l = data_preparation(LEFT_MANEUVER_FILE)

print("Preprocessing Right Maneuvers...")
l_c_r = data_preparation(RIGHT_MANEUVER_FILE)

# Print types and lengths
print(f"Type of l_c_r: {type(l_c_r)}")
print(f"Length of l_c_r: {len(l_c_r)}")
print(f"Length of l_c_l: {len(l_c_l)}")
print(f"Length of l_k: {len(l_k)}")

# Shuffle the lists
shuffle(l_k)
shuffle(l_c_r)
shuffle(l_c_l)

# Create train/test splits (assuming 120 for train, 30 for test based on user's code)
train_size = 120
test_size = 30  # Assuming total around 150 per dataset

l_c_r_Train = l_c_r[:train_size]
l_c_r_Test = l_c_r[train_size:train_size + test_size]

l_c_l_Train = l_c_l[:train_size]
l_c_l_Test = l_c_l[train_size:train_size + test_size]

l_k_Train = l_k[:train_size]
l_k_Test = l_k[train_size:train_size + test_size]

# Print split lengths
print(f"Length of l_c_r_Train: {len(l_c_r_Train)}")
print(f"Length of l_c_l_Train: {len(l_c_l_Train)}")
print(f"Length of l_k_Train: {len(l_k_Train)}")
print(f"Length of l_c_r_Test: {len(l_c_r_Test)}")
print(f"Length of l_c_l_Test: {len(l_c_l_Test)}")
print(f"Length of l_k_Test: {len(l_k_Test)}")

# Example visualization: Plot Lateral Position for Right Maneuvers (all series overlaid)
plot_series_state(right_data, state=0, title='Lane Change Right', ylab='Lateral position (m)', axis=[-1, 45, 0, 15],color='y')
# Example visualization: Plot Lateral Position for Right Maneuvers (all series overlaid)
plot_series_state(left_data, state=0, title='Lane Change Left', ylab='Lateral position (m)', axis=[-1, 45, 0, 15],color='b')
# Example visualization: Plot Lateral Position for Right Maneuvers (all series overlaid)
plot_series_state(straight_data, state=0, title='Lane Keep', ylab='Lateral position (m)', axis=[-1, 45, 0, 15])


# Custom GaussianHMM class
class GaussianHMM:
    """
    Gaussian Hidden Markov Model implemented with Baum-Welch algorithm.
    Supports multiple sequences for training.
    """
    def __init__(self, n_components, covariance_type='full', n_iter=100, tol=1e-4, random_state=42, verbose=False):
        self.n_components = n_components  ## number of hidden state (K in 1 to 10)
        self.covariance_type = covariance_type  # 'full' (full covariance matrix) or 'diag' (diagonal only, faster but assumes independent features)
        self.n_iter = n_iter # Max Expectation Maximimization iterations
        self.tol = tol       # Convergence tolerance (stop if log-likelihood improvement < tol)
        self.random_state = random_state   # For reproducible random initialization
        self.verbose = verbose   # Print progress if True

    def fit(self, seq_list):
        if not seq_list:
            raise ValueError("No sequences provided.")
        
        num_seq = len(seq_list)
        D = seq_list[0].shape[1] if isinstance(seq_list[0], np.ndarray) else seq_list[0].values.shape[1]
        
        np.random.seed(self.random_state)
        
        # Initialize parameters
        all_obs = np.concatenate([s if isinstance(s, np.ndarray) else s.values for s in seq_list])
        n_samples, _ = all_obs.shape
        
        self.startprob_ = np.ones(self.n_components) / self.n_components
        log_startprob = np.log(self.startprob_)
        
        transmat = np.random.uniform(size=(self.n_components, self.n_components))
        transmat /= transmat.sum(1, keepdims=True)
        self.transmat_ = transmat
        log_transmat = np.log(transmat)
        
        # Means: perturb around data mean
        data_mean = np.mean(all_obs, axis=0)
        data_std = np.std(all_obs, axis=0) + 1e-6
        self.means_ = data_mean + np.random.randn(self.n_components, D) * data_std
        
        # Covars: data cov + eye
        data_cov = np.cov(all_obs.T) + np.eye(D) * 1e-3
        if self.covariance_type == 'full':
            self.covars_ = np.array([data_cov for _ in range(self.n_components)])
        elif self.covariance_type == 'diag':
            self.covars_ = np.array([np.diag(np.diag(data_cov)) for _ in range(self.n_components)])
        else:
            raise ValueError("Invalid covariance_type")
        
        prev_ll = -np.inf
        
        for it in range(self.n_iter):
            total_ll = 0
            gamma_sum_0 = np.zeros(self.n_components)
            xi_sum = np.zeros((self.n_components, self.n_components))
            weighted_obs_sum = np.zeros((self.n_components, D))
            outer_dev_sum = np.zeros((self.n_components, D, D)) if self.covariance_type == 'full' else np.zeros((self.n_components, D))
            gamma_sum = np.zeros(self.n_components)
            
            for obs in seq_list:
                if isinstance(obs, pd.DataFrame):
                    obs = obs.values
                N = obs.shape[0]
                
                log_b = self._log_emission_prob(obs)
                
                # Forward
                log_alpha = np.zeros((N, self.n_components))
                log_alpha[0] = log_startprob + log_b[0]
                for t in range(1, N):
                    log_alpha[t] = logsumexp(log_alpha[t-1][:, np.newaxis] + log_transmat, axis=0) + log_b[t]
                
                seq_ll = logsumexp(log_alpha[-1])
                total_ll += seq_ll
                
                # Backward
                log_beta = np.zeros((N, self.n_components))
                log_beta[-1] = 0
                for t in range(N-2, -1, -1):
                    log_beta[t] = logsumexp(log_transmat + log_b[t+1] + log_beta[t+1], axis=1)
                
                log_gamma = log_alpha + log_beta - seq_ll
                gamma = np.exp(log_gamma)
                
                # Xi
                if N > 1:
                    log_xi = np.zeros((N-1, self.n_components, self.n_components))
                    for t in range(N-1):
                        log_xi[t] = log_alpha[t][:, np.newaxis] + log_transmat + log_b[t+1] + log_beta[t+1] - seq_ll
                    xi = np.exp(log_xi)
                    xi_sum += xi.sum(0)
                
                gamma_sum_0 += gamma[0]
                gamma_sum += gamma.sum(0)
                weighted_obs_sum += np.dot(gamma.T, obs)
                
                for k in range(self.n_components):
                    dev = obs - self.means_[k]
                    if self.covariance_type == 'full':
                        outer_dev_sum[k] += np.dot((gamma[:, k][:, np.newaxis] * dev).T, dev)
                    else:
                        outer_dev_sum[k] += np.sum(gamma[:, k][:, np.newaxis] * (dev ** 2), axis=0)
            
            # M-step
            self.startprob_ = gamma_sum_0 / num_seq
            self.startprob_ /= self.startprob_.sum()
            log_startprob = np.log(self.startprob_ + 1e-10)
            
            self.transmat_ = xi_sum / xi_sum.sum(1)[:, np.newaxis]
            self.transmat_ /= self.transmat_.sum(1)[:, np.newaxis]
            log_transmat = np.log(self.transmat_ + 1e-10)
            
            self.means_ = weighted_obs_sum / gamma_sum[:, np.newaxis]
            
            for k in range(self.n_components):
                if self.covariance_type == 'full':
                    self.covars_[k] = outer_dev_sum[k] / gamma_sum[k]
                    self.covars_[k] += np.eye(D) * 1e-6
                else:
                    self.covars_[k] = outer_dev_sum[k] / gamma_sum[k]
                    self.covars_[k] += 1e-6
            
            if self.verbose:
                print(f"Iter {it + 1}, LL: {total_ll:.2f}")
            
            if total_ll - prev_ll < self.tol:
                if self.verbose:
                    print("Converged.")
                break
            prev_ll = total_ll
        
        return self
    
    def _log_emission_prob(self, obs):
        N = obs.shape[0]
        log_b = np.zeros((N, self.n_components))
        for k in range(self.n_components):
            if self.covariance_type == 'diag':
                cov = np.diag(self.covars_[k])
            else:
                cov = self.covars_[k]
            dist = multivariate_normal(mean=self.means_[k], cov=cov, allow_singular=True)
            log_b[:, k] = dist.logpdf(obs)
        return log_b
    
    def score(self, obs):
        if isinstance(obs, pd.DataFrame):
            obs = obs.values
        N = obs.shape[0]
        log_b = self._log_emission_prob(obs)
        log_alpha = np.zeros((N, self.n_components))
        log_alpha[0] = np.log(self.startprob_ + 1e-10) + log_b[0]
        for t in range(1, N):
            log_alpha[t] = logsumexp(log_alpha[t-1][:, np.newaxis] + np.log(self.transmat_ + 1e-10), axis=0) + log_b[t]
        return logsumexp(log_alpha[-1])

def compute_bic_and_ll(train_data_list, min_states=1, max_states=15, covariance_type='full', n_iter=50, verbose=False):
    bics = []
    lls = []
    seq_list = [df.values if isinstance(df, pd.DataFrame) else df for df in train_data_list]
    all_obs = np.concatenate(seq_list)
    total_obs = all_obs.shape[0]
    D = all_obs.shape[1]
    
    for n in range(min_states, max_states + 1):
        model = GaussianHMM(n_components=n, covariance_type=covariance_type, n_iter=n_iter, verbose=verbose)
        model.fit(seq_list)
        ll = sum(model.score(seq) for seq in seq_list)
        # Number of parameters
        p_start = n - 1
        p_trans = n * (n - 1)
        p_means = n * D
        if covariance_type == 'full':
            p_cov = n * D * (D + 1) // 2
        else:
            p_cov = n * D
        p = p_start + p_trans + p_means + p_cov
        bic = -2 * ll + p * np.log(total_obs)
        bics.append(bic)
        lls.append(ll)
        print(f"States {n}: BIC = {bic:.2f}, LL = {ll:.2f}")
    
    return bics, lls

def plot_bic_and_ll(bics_dict, lls_dict):
    maneuvers = list(bics_dict.keys())
    num_maneuvers = len(maneuvers)
    fig, axs = plt.subplots(num_maneuvers, 2, figsize=(12, 4 * num_maneuvers))
    
    for idx, maneuver in enumerate(maneuvers):
        states = range(1, len(bics_dict[maneuver]) + 1)
        
        # Plot BIC
        if num_maneuvers == 1:
            axs[0].plot(states, bics_dict[maneuver], marker='o')
            axs[0].set_title(f'BIC for {maneuver}')
            axs[0].set_xlabel('Number of States')
            axs[0].set_ylabel('BIC Value')
            
            axs[1].plot(states, lls_dict[maneuver], marker='o')
            axs[1].set_title(f'Log-Likelihood for {maneuver}')
            axs[1].set_xlabel('Number of States')
            axs[1].set_ylabel('Log-Likelihood')
        else:
            axs[idx, 0].plot(states, bics_dict[maneuver], marker='o')
            axs[idx, 0].set_title(f'BIC for {maneuver}')
            axs[idx, 0].set_xlabel('Number of States')
            axs[idx, 0].set_ylabel('BIC Value')
            
            axs[idx, 1].plot(states, lls_dict[maneuver], marker='o')
            axs[idx, 1].set_title(f'Log-Likelihood for {maneuver}')
            axs[idx, 1].set_xlabel('Number of States')
            axs[idx, 1].set_ylabel('Log-Likelihood')
    
    plt.tight_layout()
    plt.show()

# Compute BIC and LL for each maneuver
print("Computing for Lane Keep...")
bic_lk, ll_lk = compute_bic_and_ll(l_k_Train, verbose=False)

print("Computing for Lane Change Left...")
bic_lcl, ll_lcl = compute_bic_and_ll(l_c_l_Train, verbose=False)

print("Computing for Lane Change Right...")
bic_lcr, ll_lcr = compute_bic_and_ll(l_c_r_Train, verbose=False)

# Store in dictionaries for plotting
bics = {
    'Lane Keep': bic_lk,
    'Lane Change Left': bic_lcl,
    'Lane Change Right': bic_lcr
}

lls = {
    'Lane Keep': ll_lk,
    'Lane Change Left': ll_lcl,
    'Lane Change Right': ll_lcr
}

# Plot the results
plot_bic_and_ll(bics, lls)

# Find optimal states (minimum BIC)
optimal_states_lk = np.argmin(bic_lk) + 1
optimal_states_lcl = np.argmin(bic_lcl) + 1
optimal_states_lcr = np.argmin(bic_lcr) + 1

print(f"Optimal states: Lane Keep = {optimal_states_lk}, LCL = {optimal_states_lcl}, LCR = {optimal_states_lcr}")

# You can override with values from plots if needed, e.g.:
# optimal_states_lk = 3
# optimal_states_lcl = 5
# optimal_states_lcr = 5

# Train models
print("Training models...")
lane_keep_model = GaussianHMM(optimal_states_lk, covariance_type='full', n_iter=100, verbose=True)
lane_keep_model.fit(l_k_Train)

lane_change_left_model = GaussianHMM(optimal_states_lcl, covariance_type='full', n_iter=100, verbose=True)
lane_change_left_model.fit(l_c_l_Train)

lane_change_right_model = GaussianHMM(optimal_states_lcr, covariance_type='full', n_iter=100, verbose=True)
lane_change_right_model.fit(l_c_r_Train)

# Save the models
pickle.dump(lane_keep_model, open('LK_HMM.bin', 'wb'))
pickle.dump(lane_change_left_model, open('LCL_HMM.bin', 'wb'))
pickle.dump(lane_change_right_model, open('LCR_HMM.bin', 'wb'))
print("Models saved.")

# Prepare labeled test data
labeled_lk_test = [[df.values, "Lane Keep"] for df in l_k_Test]
labeled_lcl_test = [[df.values, "Change Left"] for df in l_c_l_Test]
labeled_lcr_test = [[df.values, "Change Right"] for df in l_c_r_Test]

# Combine and shuffle
all_test = labeled_lk_test + labeled_lcl_test + labeled_lcr_test
shuffle(all_test)
print(f"Length of all test data: {len(all_test)}")

# Models dictionary
models = {
    "Lane Keep": lane_keep_model,
    "Change Left": lane_change_left_model,
    "Change Right": lane_change_right_model
}

# Predict and collect true/pred labels
y_true = [label for _, label in all_test]
y_pred = []
for seq, _ in all_test:
    log_probs = {label: model.score(seq) for label, model in models.items()}
    pred_label = max(log_probs, key=log_probs.get)
    y_pred.append(pred_label)

# Compute confusion matrix
labels = ["Lane Keep", "Change Left", "Change Right"]
cm = confusion_matrix(y_true, y_pred, labels=labels)

# Plot confusion matrix
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels, ax=ax)
ax.set_xlabel('Predicted')
ax.set_ylabel('Actual')
ax.set_title('HMM Classifier Confusion Matrix')
plt.show()

# Compute and print metrics
accuracy = np.trace(cm) / np.sum(cm)
precision, recall, fscore, support = precision_recall_fscore_support(y_true, y_pred, labels=labels)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {fscore}")
print(f"Support: {support}")
```

```python
# def TTC (V_Ego,X_Ego,V_s,X_s):
#     ttc=(X_s-X_Ego)/(V_Ego-V_s)
#     if ttc>0:
#         ttc=ttc
#     else:
#         ttc=100
#     return ttc

# def MSM (X_s,X_Ego):
#     msm=X_Ego-X_s
#     return msm

def two_num(num):
    count =0
    a_b=[];                              a_f=[];
    while count < num:
        Back = np.random.uniform(low=64.01   , high=128 , size=1)
        Front= np.random.uniform(low=0   , high=64  , size=1)

        if (Back-Front)>7:
            #print((Back-Front))
            #print(Back[0]-Front[0])
            a_b.append(Back )
            a_f.append(Front)
            count+=1
    return a_b ,a_f

def Data_Generator(N):
    # set a seed so that the results are consistent
    import random
    random.seed(223)
    #print(p)
    #print(sum(p))
    np.random.seed(123345)
    ##############################  Create Positions  ##########################################
    l=two_num (N)
    l_a_b  = np.array(l[0])
    l_a_f  = np.array(l[1])

    f      = np.random.uniform(low=0   , high=58 , size=N)#V

    r=two_num (N)
    r_a_b  = np.array(r[0])
    r_a_f  = np.array(r[1])

    Ego    = np.random.uniform(low=64  , high=64 , size=N)

    ##############################   Round Nombers  ###############################################
    l_a_f  = np.round(l_a_f,  decimals=1)
    l_a_b  = np.round(l_a_b,  decimals=1)
    f      = np.round(f,      decimals=1)
    r_a_b  = np.round(r_a_b,  decimals=1)
    r_a_f  = np.round(r_a_f,  decimals=1)

    Ego    = np.round(Ego,    decimals=1)

    return(l_a_f,l_a_b,f,r_a_b,r_a_f,Ego)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def print_confusion_matrix(confusion_matrix, class_names, figsize = (10,7), fontsize=14):
    """Prints a confusion matrix, as returned by sklearn.metrics.confusion_matrix, as a heatmap.

    Arguments
    ---------
    confusion_matrix: numpy.ndarray
        The numpy.ndarray object returned from a call to sklearn.metrics.confusion_matrix.
        Similarly constructed ndarrays can also be used.
    class_names: list
        An ordered list of class names, in the order they index the given confusion matrix.
    figsize: tuple
        A 2-long tuple, the first value determining the horizontal size of the ouputted figure,
        the second determining the vertical size. Defaults to (10,7).
    fontsize: int
        Font size for axes labels. Defaults to 14.

    Returns
    -------
    matplotlib.figure.Figure
        The resulting confusion matrix figure
    """
    df_cm = pd.DataFrame(
        confusion_matrix, index=class_names, columns=class_names,
    )
    fig = plt.figure(figsize=figsize)
    try:
        heatmap = sns.heatmap(df_cm, annot=True, fmt="d",annot_kws={'fontsize': 10},cmap = "Blues")
    except ValueError:
        raise ValueError("Confusion matrix values must be integers.")
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=fontsize)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=20, ha='right', fontsize=fontsize)
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    return fig
```
